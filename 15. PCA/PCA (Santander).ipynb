```python
import numpy as np
import pandas as pd
santan = pd.read_csv("F:/Kaggle/Santander Customer/train.csv")
print(santan.shape)
```

```python
santan.head()
```

```python
X = santan.iloc[:,1:370]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
Xscaled=scaler.transform(X)
Xscaled[1:5]
```

```python
from sklearn.decomposition import PCA
pca = PCA(svd_solver = 'auto')
principalComponents = pca.fit_transform(Xscaled)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
Variation = pca.explained_variance_ratio_ * 100
len(Variation)
plt.plot(list(range(0,369)),Variation)
plt.xlabel("PCs")
plt.ylabel("Scores")
plt.title('Scree Plot')
plt.show()
```

```python
Cumu_Variation =np.cumsum(pca.explained_variance_ratio_ * 100)
len(Variation)
plt.plot(list(range(0,369)),Cumu_Variation)
plt.xlabel("PCs")
plt.ylabel("Scores")
plt.title('Scree Plot')
plt.show()
```

```python
X = pd.DataFrame(principalComponents)
X.head()
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
X = X.iloc[:,0:100]
y = santan.TARGET
```

```python
type(y)
```

```python
type(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.3, 
                                                    random_state=42,
                                                    stratify=y)

model = RandomForestClassifier(random_state=1211)
y_pred = model.fit(X_train, y_train).predict(X_test)

print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Compute predicted probabilities: y_pred_prob
y_pred_prob = model.predict_proba(X_test)[:,1]

# Generate ROC curve values: fpr, tpr, thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# Plot ROC curve
import matplotlib.pyplot as plt
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

```python
roc_auc_score(y_test, y_pred_prob)
```
